# Xarray Weather Data Example\n\nThis notebook demonstrates how to use `monet-plots` to create a spatial plot from xarray's toy weather data.

```python
import xarray as xr\nimport cartopy.crs as ccrs\nfrom monet_plots.plots.spatial import SpatialPlot
```

```python
# Load the xarray toy weather dataset\nair = xr.tutorial.load_dataset('air_temperature')\nair
```

```python
# Create a spatial plot instance with map features\nplot = SpatialPlot(\n    projection=ccrs.PlateCarree(), \n    figsize=(12, 8),\n    coastlines=True,\n    borders=True,\n    gridlines=True\n)\n\n# Select a time slice and plot the data\nair_slice = air.air.isel(time=0)\nplot.plot(air_slice, transform=ccrs.PlateCarree())\n\n# Add a title\nplot.ax.set_title('Surface Air Temperature')\n\n# Save the plot\nplot.save('xarray_weather_example.png')
```
