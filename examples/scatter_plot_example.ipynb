# Scatter Plot Example

This notebook demonstrates how to use `monet-plots` to create a scatter plot.

```python
import numpy as np
import pandas as pd
from monet_plots.plots.scatter import ScatterPlot
import matplotlib.pyplot as plt
```

```python
# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100) + np.random.rand(100) * 0.2
})
```

```python
# Create a ScatterPlot instance
plot = ScatterPlot(figsize=(8, 6))

# Plot the data
plot.plot(data, x='x', y='y', label='Sample Data')

# Add title and legend
plot.ax.set_title('Scatter Plot Example')
plot.ax.set_xlabel('X-axis')
plot.ax.set_ylabel('Y-axis')
plot.ax.legend()

# Save the plot
plot.save('scatter_plot_example.png')
```
