# Taylor Diagram Plot Example\n\nThis notebook demonstrates how to use `monet-plots` to create a Taylor Diagram.

```python
import numpy as np\nimport pandas as pd\nfrom monet_plots.taylordiagram import TaylorDiagram\nimport matplotlib.pyplot as plt
```

```python
# Generate sample data\nnp.random.seed(42)\nobs = np.random.rand(100)\nmodel1 = obs + np.random.rand(100) * 0.1\nmodel2 = obs + np.random.rand(100) * 0.2 - 0.1\n\ndf1 = pd.DataFrame({'obs': obs, 'model': model1})\ndf2 = pd.DataFrame({'obs': obs, 'model': model2})
```

```python
# Create a TaylorDiagram instance\n# The reference standard deviation is typically the standard deviation of the observations\nref_std = obs.std()\nplot = TaylorDiagram(ref_std, fig=plt.figure(figsize=(9, 8)))\n\n# Add samples to the diagram\nstddev1 = df1['model'].std()\ncorrcoef1 = df1['model'].corr(df1['obs'])\nplot.add_sample(stddev1, corrcoef1, label='Model 1', marker='o', ms=10, ls='', mfc='blue', mec='blue')\n\nstddev2 = df2['model'].std()\ncorrcoef2 = df2['model'].corr(df2['obs'])\nplot.add_sample(stddev2, corrcoef2, label='Model 2', marker='s', ms=10, ls='', mfc='red', mec='red')\n\n# Add a legend\nplot.add_legend()\n\n# Add a title\nplot.fig.suptitle('Taylor Diagram Example', fontsize=16)\n\n# Save the plot\nplot.fig.savefig('taylor_diagram_plot_example.png')
```
