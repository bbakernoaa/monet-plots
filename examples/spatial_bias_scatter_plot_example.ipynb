# Spatial Bias Scatter Plot ExampleThis notebook demonstrates how to use `monet-plots` to create a spatial bias scatter plot.

```python
import numpy as npimport pandas as pdimport cartopy.crs as ccrsimport cartopy.feature as cfeaturefrom monet_plots.plots.spatial_bias_scatter import SpatialBiasScatterPlotimport matplotlib.pyplot as plt
```

```python
# Generate sample datanp.random.seed(42)n_points = 100lon = np.random.uniform(-10, 10, n_points)lat = np.random.uniform(20, 40, n_points)obs = np.random.rand(n_points) * 10model = obs + np.random.rand(n_points) * 2 - 1 # Model with some biasdata = pd.DataFrame({    'lon': lon,    'lat': lat,    'obs': obs,    'model': model})
```

```python
# Create a SpatialBiasScatterPlot instanceplot = SpatialBiasScatterPlot(figsize=(10, 8), projection=ccrs.PlateCarree())# Plot the dataplot.plot(data, x='lon', y='lat', obs='obs', model='model', cmap='RdBu_r', vmin=-5, vmax=5)# Add geographical featuresplot.ax.add_feature(cfeature.COASTLINE)plot.ax.add_feature(cfeature.BORDERS, linestyle=':')plot.ax.add_feature(cfeature.LAND, facecolor='lightgray')plot.ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)# Add titleplot.ax.set_title('Spatial Bias Scatter Plot Example')# Save the plotplot.save('spatial_bias_scatter_plot_example.png')
```
