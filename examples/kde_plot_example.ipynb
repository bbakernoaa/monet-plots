# KDE Plot Example

This notebook demonstrates how to use `monet-plots` to create a Kernel Density Estimate (KDE) plot.

```python
import numpy as np
import pandas as pd
from monet_plots.plots.kde import KDEPlot
import matplotlib.pyplot as plt
```

```python
# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
# Create a KDEPlot instance
plot = KDEPlot(figsize=(8, 6))

# Plot the data
plot.plot(data, label='Sample Data')

# Add title and legend
plot.ax.set_title('Kernel Density Estimate Plot')
plot.ax.set_xlabel('Value')
plot.ax.set_ylabel('Density')
plot.ax.legend()

# Save the plot
plot.save('kde_plot_example.png')
```
